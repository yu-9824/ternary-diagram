```python
from ternary_diagram import TernaryDiagram
import pandas as pd
```

```python
df = pd.read_csv('example_scatter.csv')
```

```python
materials = df.columns[0:3]
```

```python
td = TernaryDiagram(materials=materials)
td.scatter(df[materials], z = df['z'], minimum = 0, maximum = 1)
```

```python
td.fig.savefig('example_scatter.png', dpi = 300)
```
